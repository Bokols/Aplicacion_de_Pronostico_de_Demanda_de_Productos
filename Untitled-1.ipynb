```python
def get_customer_id_from_invoice(invoiceno):
    """
    Extracts a customer ID from the invoice number.
    Assumes that the first 3 digits of the invoice number correspond to the customer ID.
    
    Args:
    invoice_number (str or int): The invoice number from which to extract the customer ID.

    Returns:
    str: The inferred customer ID based on the invoice number.
    """
    # Ensure the invoice_number is a string
    invoice_str = str(invoiceno)

    # If the invoice number has fewer than 3 digits, handle it as a special case
    if len(invoice_str) < 6:
        return "Unknown"  # Default value for invoices with fewer than 3 digits

    # Extract the first 3 digits and use it as the customer ID
    customer_id = invoice_str[:6]  # First 3 digits as customer ID

    return customer_id
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sentence_transformers import SentenceTransformer
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm  # Import tqdm for progress bar

# Initialize the Sentence-BERT model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Function to encode descriptions in parallel
def encode_descriptions(batch):
    return model.encode(batch)

# Function to process the embeddings and store them in the DataFrame
def process_embeddings(df):
    descriptions = df['description'].astype(str).tolist()
    batch_size = 256
    batches = [descriptions[i:i + batch_size] for i in range(0, len(descriptions), batch_size)]
    embeddings = []
    
    # Use tqdm to show progress for batch encoding
    with ThreadPoolExecutor(max_workers=4) as executor:
        results = executor.map(encode_descriptions, batches)
        for result in tqdm(results, desc="Encoding Descriptions", total=len(batches)):
            embeddings.extend(result)

    df['embeddings'] = embeddings
    return df

# Process the embeddings
df = process_embeddings(df)

# Convert embeddings to NumPy array for KMeans
X = np.vstack(df['embeddings'].values)

# Initialize the MiniBatchKMeans model
inertia = []
cluster_range = range(2, 15)  # Test 2 to 25 clusters

# Loop over different cluster sizes and calculate inertia with a progress bar
for k in tqdm(cluster_range, desc="Clustering", unit="cluster"):
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=1000)  # Using smaller batches
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o', linestyle='-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
from tqdm import tqdm  # Progress bar

# Assuming 'df' is your original DataFrame, and 'description' contains the text data

# Step 1: Convert text data into numerical features using TF-IDF
print("Computing TF-IDF features...")
tfidf_vectorizer = TfidfVectorizer(stop_words='english')  # Removing common English stopwords
X_tfidf = tfidf_vectorizer.fit_transform(tqdm(df['description'], desc="Processing Text"))

# Step 2: Fit KMeans with the optimum number of clusters (2342)
print("Clustering data using KMeans...")
kmeans = KMeans(n_clusters=2342, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_tfidf)

# Step 3: Get centroids for each cluster
centroids = kmeans.cluster_centers_

# Step 4: Generate cluster names based on the most important words in each cluster centroid
# We will focus on the top 3 most important words for each cluster

def generate_cluster_name(centroid, feature_names, top_n=3):
    """Generate a descriptive name for a cluster based on the top words from its centroid."""
    top_features_idx = np.argsort(centroid)[::-1][:top_n]  # Get indices of top N features
    top_features = [feature_names[i] for i in top_features_idx]  # Extract feature names
    return "_".join(top_features)  # Join top words to form cluster name

# Step 5: Get the feature names (terms used in TF-IDF)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Step 6: Generate and assign names to clusters with a progress bar
cluster_names = {}

print("Generating cluster names...")
for i, centroid in tqdm(enumerate(centroids), total=len(centroids), desc="Naming Clusters"):
    cluster_name = generate_cluster_name(centroid, feature_names)
    cluster_names[i] = cluster_name

# Step 7: Assign generated names to clusters
df['cluster_name'] = df['cluster'].map(cluster_names)

# Print out the first few rows to verify
print(df[['description', 'cluster', 'cluster_name']].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import MiniBatchKMeans
import matplotlib.pyplot as plt
from tqdm import tqdm
from kneed import KneeLocator
from scipy.sparse import csr_matrix


# One-Hot Encode (Sparse)
encoder = OneHotEncoder(sparse_output=True, drop='first')
cluster_name_encoded = encoder.fit_transform(df[['cluster_name']])  

# Convert to CSR format for efficient computation
cluster_name_encoded = csr_matrix(cluster_name_encoded)

# Step 2: Reduce Dimensionality with TruncatedSVD (Using a reduced number of components)
n_components = min(30, cluster_name_encoded.shape[1])  # Reduce to 30 or fewer components
svd = TruncatedSVD(n_components=n_components, random_state=42)

# Transform only once
X_reduced = svd.fit_transform(cluster_name_encoded)

print(f"Reduced shape: {X_reduced.shape}")

# Step 3: Determine Optimal Clusters (Elbow Method)
wcss = []  # Within-Cluster Sum of Squares

# Testing different k values
k_values = range(2, 15)  # Adjust range as needed

for k in tqdm(k_values, desc="Finding Optimal K"):
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=1024, n_init=10)
    kmeans.fit(X_reduced)
    wcss.append(kmeans.inertia_)

# Find the elbow point
knee = KneeLocator(k_values, wcss, curve="convex", direction="decreasing")
optimal_k = knee.elbow

# Print the most optimal number of clusters
print(f"The most optimal number of clusters is: {optimal_k}")

# Step 4: Apply Clustering with Optimal K
kmeans_final = MiniBatchKMeans(n_clusters=optimal_k, random_state=42, batch_size=1024, n_init=10)
clusters = kmeans_final.fit_predict(X_reduced)

# Add cluster labels to the DataFrame
df['cluster'] = clusters

# Step 5: Plot Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import MiniBatchKMeans, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np
import matplotlib.pyplot as plt
from joblib import Parallel, delayed
from tqdm import tqdm

# Step 1: One-Hot Encoding (Memory-Efficient)
X_cluster_name = df[['cluster_name']]
encoder = OneHotEncoder(sparse_output=True, drop='first')
cluster_name_encoded = encoder.fit_transform(X_cluster_name)

# Step 2: Dimensionality Reduction with TruncatedSVD (Directly with Sparse Matrices)
svd = TruncatedSVD(n_components=20)
cluster_name_encoded_reduced = svd.fit_transform(cluster_name_encoded)

# Step 3: Scaling
scaler = StandardScaler()
cluster_name_encoded_reduced_scaled = scaler.fit_transform(cluster_name_encoded_reduced)

# Step 4: Function to Compute Clustering Labels
def compute_kmeans_labels(k, X_data):
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, n_init=10, max_iter=300, batch_size=1000)
    return kmeans.fit_predict(X_data)

# Step 5: Parallel KMeans Clustering with Reduced k_values Range
k_values = range(2, 10)  # Reduced from (2, 10) to minimize memory usage
labels_list = Parallel(n_jobs=-1)(
    delayed(compute_kmeans_labels)(k, cluster_name_encoded_reduced_scaled)
    for k in tqdm(k_values, desc="KMeans Clustering", unit="cluster")
)

# Step 6: Function for Silhouette Score Calculation (Subset for Efficiency)
def compute_silhouette_score(X, labels, sample_size=5000):
    if len(X) > sample_size:
        indices = np.random.choice(len(X), size=sample_size, replace=False)
        X_sampled = X[indices]
        labels_sampled = np.array(labels)[indices]
    else:
        X_sampled = X
        labels_sampled = labels
    return silhouette_score(X_sampled, labels_sampled, metric='euclidean')

# Step 7: Silhouette Scores Calculation
silhouette_scores_kmeans = []
for labels in tqdm(labels_list, desc="Calculating Silhouette Scores", unit="score"):
    score = compute_silhouette_score(cluster_name_encoded_reduced_scaled, labels)
    silhouette_scores_kmeans.append(score)

# Step 8: Plot Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(k_values, silhouette_scores_kmeans, marker='o', label='KMeans')
plt.title("Silhouette Scores for Different k Values (KMeans)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.legend()
plt.show()

# Step 9: Optimal K Selection
optimal_k = k_values[np.argmax(silhouette_scores_kmeans)]
print(f"Optimal number of clusters for KMeans: {optimal_k}")
```

```python
# Compute IQR for each numerical column
outlier_dict = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)  # First quartile (25%)
    Q3 = df[col].quantile(0.75)  # Third quartile (75%)
    IQR = Q3 - Q1  # Interquartile Range
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    outlier_dict[col] = {
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": len(outliers),
        "Outlier Percentage": (len(outliers) / len(df)) * 100
    }

# Convert outlier dictionary to dataframe for a summary table
outlier_summary = pd.DataFrame(outlier_dict).T
print(outlier_summary)
```

```python
# Make sure df_cleaned is a copy to avoid the warning
df = df.copy()

# Use .loc to avoid SettingWithCopyWarning
df.loc[:, 'lag_1'] = df['quantity'].shift(1)
df.loc[:, 'lag_7'] = df['quantity'].shift(7)
```
