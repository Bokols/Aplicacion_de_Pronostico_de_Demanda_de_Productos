```python
def get_customer_id_from_invoice(invoiceno):
    """
    Extracts a customer ID from the invoice number.
    Assumes that the first 3 digits of the invoice number correspond to the customer ID.
    
    Args:
    invoice_number (str or int): The invoice number from which to extract the customer ID.

    Returns:
    str: The inferred customer ID based on the invoice number.
    """
    # Ensure the invoice_number is a string
    invoice_str = str(invoiceno)

    # If the invoice number has fewer than 3 digits, handle it as a special case
    if len(invoice_str) < 6:
        return "Unknown"  # Default value for invoices with fewer than 3 digits

    # Extract the first 3 digits and use it as the customer ID
    customer_id = invoice_str[:6]  # First 3 digits as customer ID

    return customer_id
```

```python
# Example: Fill missing customerid based on invoice_number pattern (simple example)
df['customerid'] = df.apply(
    lambda row: get_customer_id_from_invoice(row['invoiceno']) if pd.isna(row['customerid']) else row['customerid'],
    axis=1
)
```

```python
# Compute IQR for each numerical column
outlier_dict = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)  # First quartile (25%)
    Q3 = df[col].quantile(0.75)  # Third quartile (75%)
    IQR = Q3 - Q1  # Interquartile Range
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    outlier_dict[col] = {
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": len(outliers),
        "Outlier Percentage": (len(outliers) / len(df)) * 100
    }

# Convert outlier dictionary to dataframe for a summary table
outlier_summary = pd.DataFrame(outlier_dict).T
print(outlier_summary)
```

```python
# Remove outliers for each numerical column using IQR
for col in numerical_cols:
    # Calculate Q1, Q3, and IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter out rows where the column values are outside the bounds
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Display the cleaned dataframe
print(df.shape)  # Print the shape of the dataframe after removing outliers
print(df.head())  # Print the first few rows of the cleaned dataframe
```

# Feature Engineering

Time-based Features: Create additional time-based features to capture seasonality, trends, and periodicity. Some common time-related features for demand forecasting are:

Day of the week

Month of the year

Year

Is it a weekend or holiday? (This is important for retail and transportation industries)

```python
df.head()
```

```python
df['day_of_week'] = df['invoicedate'].dt.dayofweek
df['month'] = df['invoicedate'].dt.month
df['is_weekend'] = df['invoicedate'].dt.weekday >= 5  # Saturday or Sunday
```

Lag Features: In time-series data, past observations often help predict future ones. Lag features are previous values in the time series, which can be used as predictors. For example, last week’s sales can help predict this week’s sales.

```python
df['lag_1'] = df['quantity'].shift(1)
df['lag_7'] = df['quantity'].shift(7)
```

Rolling Features: These features provide insights into the trend or seasonality by calculating moving averages or sums over a specific window of time.

```python
df['rolling_mean_7'] = df['quantity'].rolling(window=7).mean()
df['rolling_std_7'] = df['quantity'].rolling(window=7).std()
```

Holiday and Event Features: Add information about holidays, special events, or promotions, which can have a significant impact on demand.

```python
# Dictionary to map full country names to abbreviations
country_abbreviations = {
    'United Kingdom': 'UK',
    'France': 'FRA',
    'Australia': 'AU',
    'Netherlands': 'NL',
    'Germany': 'DE',
    'Norway': 'NO',
    'EIRE': 'IE',
    'Switzerland': 'CH',
    'Belgium': 'BE',
    'Spain': 'ES',
    'Poland': 'PL',
    'Portugal': 'PT',
    'Italy': 'IT',
    'Lithuania': 'LT',
    'Japan': 'JP',
    'Iceland': 'DK',
    'Channel Islands': 'UK',
    'Denmark': 'DK',
    'Cyprus': 'UK',
    'Sweden': 'SE',
    'Austria': 'AT',
    'Israel': 'IS',
    'Finland': 'FI',
    'Bahrain': 'BH',
    'Greece': 'GR',
    'Hong Kong': 'HK',
    'Singapore': 'SI',
    'Lebanon': 'LB',
    'United Arab Emirates': 'UAE',
    'Saudi Arabia': 'KSA',
    'Czech Republic': 'CZ',
    'Canada': 'CA',
    'Unspecified': 'N/A',
    'Brazil': 'BR',
    'European Community': 'EU',
    'Malta': 'MT',
    'RSA': 'ZA',
    'USA': 'US'
}

# Function to create a new column with country abbreviations
def add_country_abbreviation(df, column_name, mapping_dict):
    # Create a new column with the mapped abbreviations
    df['countryabbreviation'] = df[column_name].map(mapping_dict)
    return df

# Apply the function to add the abbreviation column
df = add_country_abbreviation(df, 'country', country_abbreviations)

print(df)
```

```python
import holidays

#Convert country names



# Function to check if a date is a holiday in a given country
def is_holiday(date, country):
    try:
        # Check if the country exists in the holidays library
        country_holidays = holidays.CountryHoliday(country)
        return date in country_holidays
    except NotImplementedError:
        # If the country code is invalid, return False
        return False

# Apply the function to create a new 'is_holiday' column
df['is_holiday'] = df.apply(lambda row: is_holiday(row['invoicedate'], row['countryabbreviation']), axis=1)

# Display the DataFrame
print(df)
```

4. Time Series Analysis
Since this is a demand forecasting task, understanding temporal trends is key.

a. Plotting Time Series of Quantity Sold

```python
# Resample by day and sum the quantities sold
daily_sales = df.resample('D', on='invoicedate')['quantity'].sum()

plt.figure(figsize=(12, 6))
daily_sales.plot()
plt.title('Daily Quantity Sold')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.show()
```

b. Autocorrelation Plot (ACF and PACF)

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.figure(figsize=(12, 6))

# ACF plot
plt.subplot(1, 2, 1)
plot_acf(daily_sales, lags=50, ax=plt.gca())
plt.title('ACF (Autocorrelation) Plot')

# PACF plot
plt.subplot(1, 2, 2)
plot_pacf(daily_sales, lags=50, ax=plt.gca())
plt.title('PACF (Partial Autocorrelation) Plot')

plt.tight_layout()
plt.show()
```

c. Seasonal Decomposition of Time Series (STL)

```python
from statsmodels.tsa.seasonal import STL

# Apply STL decomposition
stl = STL(daily_sales, seasonal=13)
result = stl.fit()

# Plot the decomposition
result.plot()
plt.show()
```
